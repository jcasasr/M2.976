# Càlcul de les mètriques bàsiques de nodes

## Creació dels grafs

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.lollipop_graph(6, 4)

nx.draw(G, with_labels=True)
plt.show()
```

```python
# Un de dirigit

D = nx.gn_graph(10, seed = 0)
nx.draw(D, with_labels=True)
plt.show()
```

## Grau

```python
# Undirected

print("Grau de 8: %d" % nx.degree(G,8))
print("Grau de 2: %d" % nx.degree(G,2))
print("Grau de 5: %s" % nx.degree(G,5))
```

```python
# Directed. Observeu que canvia la manera de fer la crida a la funció.

print("InDegree de 1: %d" % D.in_degree(1))
print("OutDegree de 1: %d" % D.out_degree(1))
print("Degree de 1: %s" % D.degree(1))
```

```python
# Histograma de graus (van ordenats en grau 0, 1, 2, 3...).

print(nx.degree_histogram(G))
print(nx.degree_histogram(D))
```

```python
# Veïns (adjacents) al node 5.
list(nx.neighbors(G,5))
```

```python
# En un dirigit, els veïns només són de sortida.
list(nx.neighbors(D,1))
```

```python
# I així s'obtenen tots els «contigus».
list(nx.all_neighbors(D,1))
```
